```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('VGE_FGM_dataset.csv')
df.count()
```

```python
df = df.dropna()
df.count()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.loc[df['Immediate needs on arrival/Clothes']==1,"Immediate needs on arrival"] = 'Clothes'
df.loc[df['Immediate needs on arrival/Underwear']==1,"Immediate needs on arrival"] = 'Underwear'
df.loc[df['Immediate needs on arrival/Shoes']==1,"Immediate needs on arrival"] = 'Shoes'
df.loc[df['Immediate needs on arrival/Counselling']==1,"Immediate needs on arrival"] = 'Counselling'
df.loc[df['Immediate needs on arrival/Other']==1,"Immediate needs on arrival"] = 'Other'
df.loc[df['Health Status/Healthy']==1,"Health Status"] = 'Heathy'
df.loc[df['Health Status/STI']==1,"Health Status"] = 'STI'
df.loc[df['Health Status/UTI']==1,"Health Status"] = 'UTI'
df.loc[df['Health Status/Pregnant']==1,"Health Status"] = 'Pregnant'
df.loc[df['Health Status/Other health issue']==1,"Health Status"] = 'Other Issues'
```

```python
df
```

```python
df['Date of arrival at Safe House'] = pd.to_datetime(df['Date of arrival at Safe House'])
```

```python
by_Region = df.groupby('Region')
by_Region.first()
```

```python
print(df.groupby('Region').size())
```

```python
by_Region.District.value_counts()
```

```python
by_School = df.groupby('Do you go to school?')
by_School.first()
```

```python
df['School'] = df['Do you go to school?'].transform(lambda x: 0 if x == 'no, never went' else 1)
```

```python
print(df.groupby('School').size())
```

```python
df1 = df.drop(['Do you go to school?','Immediate needs on arrival/Clothes','Immediate needs on arrival/Underwear','Immediate needs on arrival/Shoes',
               'Immediate needs on arrival/Counselling','Immediate needs on arrival/Other',
              'Health Status/Healthy','Health Status/STI','Health Status/UTI','Health Status/Pregnant','Health Status/Other health issue'], axis=1)
```

```python
data_with_index = df1.set_index("Region")
data_with_index.head()
```

```python
df2 = data_with_index.drop(index=(['Dodoma','Kilimanjaro','Manyara','Singida']))
```

```python
df2.to_csv('Safe Houses for Girls and Women.csv')
```
